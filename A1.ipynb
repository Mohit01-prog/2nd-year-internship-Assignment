```python
name = input("Enter name of the student: ")
student_class = input("Enter class: ")

marks1 = float(input("Enter marks for subject 1: "))
marks2 = float(input("Enter marks for Subject 2: "))
marks3 = float(input("Enter marks for Subject 3: "))
marks4 = float(input("Enter marks for Subject 4: "))
marks5 = float(input("Enter marks for Subject 5: "))

total_marks = marks1 + marks2 + marks3 + marks4 + marks5
percentage = (total_marks / 500) * 100

print("\n------ Student Result ------")
print("Student Name :", name)
print("Class        :", student_class)
print("Total Marks  :", total_marks)
print("Percentage   :", percentage, "%")
```

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

result = str1 + " " + str2

print("\nConcatenated String:", result)

print("\n----- String Methods -----")

print("Lower Case:", result.lower())

print("Upper Case:", result.upper())

print("Title Case:", result.title())

print("Swap Case:", result.swapcase())

print("Capitalize:", result.capitalize())

print("Casefold:", result.casefold())

print("Centered String:", result.center(50, "*"))

print("Count of letter 'a':", result.count('a'))

print("Ends with 'a':", result.endswith('a'))

print("Find word 'hello':", result.find('hello'))

print("Is Alphanumeric:", result.isalnum())

print("Is Digit:", result.isdigit())

print("Is Numeric:", result.isnumeric())

print("Is Space:", result.isspace())

print("Replace word:", result.replace("hello", "hi"))
```

```python
a = 10
b = 5

print("Initial value of a:", a)

a += b
print("After += :", a)

a -= b
print("After -= :", a)

a *= b
print("After *= :", a)

a /= b
print("After /= :", a)

a %= b
print("After %= :", a)

a **= 2
print("After **= :", a)

a //= 2
print("After //= :", a)
```

```python
name = input("Enter student name: ")
student_class = input("Enter class: ")

mark1 = float(input("Enter marks of Subject 1: "))
mark2 = float(input("Enter marks of Subject 2: "))
mark3 = float(input("Enter marks of Subject 3: "))
mark4 = float(input("Enter marks of Subject 4: "))
mark5 = float(input("Enter marks of Subject 5: "))

total = mark1 + mark2 + mark3 + mark4 + mark5
percentage = (total / 500) * 100

if percentage >= 60:
    grade = 'A'

elif percentage >= 50 and percentage < 60:
    grade = 'B'

elif percentage >= 40 and percentage < 50:
    grade = 'C'

elif percentage >= 33 and percentage < 40:
    grade = 'D'

else:
    grade = 'Fail'

print("\n------ Student Result ------")
print("Student Name :", name)
print("Class :", student_class)
print("Total Marks :", total)

print("Percentage :", percentage, "%")

print("Grade :", grade)
```
