#### 1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]] 

```python
array = np.array([[6, -8, 73, -110], [np.nan, -8, 0, 94]] )
print("Array: \n",array)
replace_nan = np.nan_to_num(array,nan = 0)
print("\nAfter replacing nan with zero : \n",replace_nan)
interchange = replace_nan.T
print("\nAfter interchange : \n",interchange)

# it is only a 2 X 4 array so interchange of 3 rows and three columns is not possible but if there is a higher order array is given like 6 X 6 array then we can use slicing like this - 
# sub = arr[:3,:3]
# arr[:3,:3] = sub.T 
# to achieve the interchange of specified rows and columns
```

#### 2) Move axes of 3D array to new positions 

```python
arr = np.arange(24).reshape(2, 3, 4)

print("\nOriginal 3D Array Shape:", arr.shape)
print(arr)

moved = np.moveaxis(arr, 0, 2)

print("\nAfter moving axis 0 to axis 2:")
print("New Shape:", moved.shape)
print(moved)
print("\n")
moved = np.moveaxis(arr, 0, 1)

print("\nAfter moving axis 0 to axis 1:")
print("New Shape:", moved.shape)
print(moved)
```

#### 3) Replace NaN values with average of columns 

```python

arr = np.array([[10, 20, np.nan],
                [30, np.nan, 50],
                [40, 60, 70]])

print("\nOriginal Array with NaN:\n", arr)

col_mean = np.nanmean(arr, axis=0)

print("\nColumn Means:\n", col_mean)

inds = np.where(np.isnan(arr))
print("\n",inds)
arr[inds] = np.take(col_mean, inds[1])

print("\nAfter replacing NaN with column averages:\n", arr)
```

#### 4) Replace negative value with zero in numpy array using replace

```python
arr = np.array([[4, -2, 7],[-5, 9, -1]])

print("\nOriginal Array:\n", arr)

arr[arr < 0] = 0

print("\nAfter replacing negative values with 0:\n", arr)
```
