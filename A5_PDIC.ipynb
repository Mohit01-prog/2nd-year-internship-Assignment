#### 2) Write a Python program to join three separate tuples into one new tuple using the + operator. 

```python
t1 = (1,2,3)
t2 = (4,5,6)
t3 = (7,8,9)
print(f"""
{t1}
{t2}
{t3}""")
joined_t = t1 + t2 + t3
print(joined_t,type(joined_t))
```

#### 4) Write a Python program to calculate the total, highest value, and lowest value from a tuple of integers without using the built-in sum (), max(), and min() functions. 

```python
t = (1,2,3,4,9,8,7,6,5)
sum = 0
max = t[0]
min = t[0]
for i in t:
    sum += i
    if max < i: max = i
    if min > i: min = i
print("total:",sum)
print("max:",max)
print("min:",min)    
```

#### 5) Write a Python program to filter a tuple . n = (3, 14, 7, 22, 9, 41, 18, 5), keep only values greater than 10 

```python
n = (3,14,7,22,9,41,18,5)
print("given tuple: ",n)
filtered_n = tuple(x for x in n if x>10)
print("filtered tuple(with only values > 10): ",filtered_n)
```

#### 7) Write a Python program to combine two sets into one, containing all unique elements from both sets. 

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7,8}
s_unique = s1 | s2
print("s1:",s1)
print("s2:",s2)
print("combined:",s_unique) 
```

#### 8) Write a Python program to find all elements that are common to both sets. s1 = {1, 2, 3, 4} s2 = {3, 4, 5, 6} 

```python
s1 = {1,2,3,4}
s2 = {3,4,5,6}
s_common = s1 & s2
print("s1:",s1)
print("s2:",s2)
print("common:",s_common) 
```
