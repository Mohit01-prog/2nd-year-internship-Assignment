#1) Explore more regex patterns Eg. The regex pattern used to validate email addresses, mobile no, string, and more 

```python


import pandas as pd

data ={
    "text":[
        "apple123",
        "banana456",
        "cat789",
        "dog"
    ]
}

df = pd.DataFrame(data)
print(df,"\n")
df["has_digit"] = df["text"].str.contains(r"\d")
print(df,"\n")
df["match"] = df["text"].str.match(r"cat")
print(df,"\n")
df["fullmatch"] = df["text"].str.fullmatch(r"[a-z]+\d+")
print(df,"\n")
df[["letter","number"]] = df["text"].str.extract(r"([A-Za-z]+)(\d+)")
print(df,'\n')
res = df["text"].str.extractall(r"([a-z])(\d)")
print(res,"\n")
df["no_digit"] = df["text"].str.replace(r"\d+","",regex = True)
print(df,"\n")
df["all_digit"] = df["text"].str.findall(r"\d")
print(df,"\n")
df["vowel_count"] = df["text"].str.count(r"[aeiouAEIOU]")
print(df,"\n")
df["split_data"] = df["text"].str.split(r"\d+")
print(df,"\n")
```

```python
data = {
    "text1":[
        "emily123@gmail.com",
        "dog",
        "hello1@yahoo.com"
    ]
}
df1 = pd.DataFrame(data)
df1["email"] = df1["text1"].str.extract(r"([\w\.-]+@[\w\.-]+\.\w+)")
print(df1)
```

#2) Explore more datetime function and uses in pandas

```python
import pandas as pd

df = pd.DataFrame({
    "date": [
        "2025-01-10",
        "2025-02-15",
        "2025-03-20"
    ],
    "sales": [100, 200, 300]
})

df["date"] = pd.to_datetime(df["date"])

df["year"] = df["date"].dt.year

df["month"] = df["date"].dt.month_name()

df["day"] = df["date"].dt.day_name()

df["next_week"] = df["date"] + pd.Timedelta(days=7)

df["formatted_date"] = df["date"].dt.strftime("%d-%m-%Y")

print(df)
```

#3) Import an meaningful csv file for data analysis and perform data cleaning, and analysis for meaningful output

```python
import pandas as pd

df = pd.read_csv("tips.csv")

print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())

df = df.drop_duplicates()

print("\nAverage Tip:")
print(df["tip"].mean())

print("\nMaximum Tip:")
print(df["tip"].max())

print("\nDay-wise Total Bill:")
print(df.groupby("day")["total_bill"].sum())

print("\nSmoker vs Non-Smoker Average Tip:")
print(df.groupby("smoker")["tip"].mean())
```
