#### 2)Create two DataFrames, df1 and df2, with a common column (e.g., 'ID').

###### > Perform an inner merge on this common column and display the resulting DataFrame.

###### > Perform a left join of df1 and df2 on the 'ID' column. Explain how missing values are handled in the resulting DataFrame. Right Join and Index-Based Join.

###### > Perform a right join using pd.merge() on a common column, then perform a join using df.join() based on the index. Compare the results. Merging with Multiple Keys.

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aman', 'Rohit', 'Simran', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4, 5],
    'Marks': [85, 90, 88, 95]
})

print("DF1")
print(df1)

print("\nDF2")
print(df2)
```

```python
inner_merge = pd.merge(df1, df2, on='ID', how='inner')

print(inner_merge)
```

```python
left_join = pd.merge(df1, df2, on='ID', how='left')

print(left_join)
```

```python
right_join = pd.merge(df1, df2, on='ID', how='right')

print(right_join)
```

```python
df1_index = df1.set_index('ID')
df2_index = df2.set_index('ID')

joined_df = df1_index.join(df2_index)

print(joined_df)
```

```python
df3 = pd.DataFrame({
    'ID': [1, 1, 2, 2],
    'Subject': ['Math', 'Science', 'Math', 'Science'],
    'Marks': [90, 85, 88, 92]
})

df4 = pd.DataFrame({
    'ID': [1, 2, 2],
    'Subject': ['Math', 'Math', 'Science'],
    'Teacher': ['Raj', 'Anita', 'Kiran']
})

multi_merge = pd.merge(df3, df4, on=['ID', 'Subject'])

print(multi_merge)
```

#### 3) Create three DataFrames. Vertically concatenate two of them using pd.concat(), then merge the resulting DataFrame with the third DataFrame on a common key. T Understand join() vs. merge().Also Explain the primary differences between df.join() and pd.merge(),

```python
df_a = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Rohit']
})

df_b = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Simran', 'Neha']
})

df_c = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'City': ['Jaipur', 'Delhi', 'Mumbai', 'Pune']
})

print(df_a)
print(df_b)
print(df_c)

concat_df = pd.concat([df_a, df_b], axis=0)

print(concat_df)

final_merge = pd.merge(concat_df, df_c, on='ID')

print(final_merge)
```

| Basis               | `join()`             | `merge()`             |
| ------------------- | -------------------- | --------------------- |
| Simplicity          | Easier               | More detailed         |
| Default Join        | Left Join            | Inner Join            |
| Based On            | Index                | Columns               |
| Syntax              | Short                | More flexible         |
| Multiple Keys       | Difficult            | Easy                  |
| SQL-like Operations | Limited              | Excellent             |
| Performance         | Good for index joins | Good for column joins |
