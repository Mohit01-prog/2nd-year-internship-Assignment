```python
#Write a Python function to find the maximum of three numbers.\

def function(a,b,c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    else:
        return c

a = int(input("Enter First number"))
b = int(input("Enter Second number"))
c = int(input("Enter Third number"))

result = function(a,b,c)

print(result)
```

```python
#Write a Python function that takes a list and returns a new list with distinct elements from the first list.

def distinct_list(items):
    
    unique_list = []
    
    for item in items:
        
        if item not in unique_list:
            unique_list.append(item)
    
    return unique_list

numbers = input("Enter elements: ")

numbers_list = numbers.split()

result = distinct_list(numbers_list)

print("List with distinct elements:", result)
```

```python
#Write a Python function to multiply all the numbers in a list.

def multiply_list(numbers):
    
    result = 1
    
    for num in numbers:
        result = result * num
    
    return result

nums = input("Enter numbers: ")

temp_list = nums.split()

numbers_list = []

for i in temp_list:
    numbers_list.append(int(i))

product = multiply_list(numbers_list)

print("Multiplication of all numbers is:", product)
```

```python
#Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

def factorial(n):
    
    fact = 1
    
    for i in range(1, n + 1):
        fact = fact * i
    
    return fact

num = int(input("Enter a number: "))

if num < 0:
    print("Factorial does not exist.")

else:
    result = factorial(num)
    print("Factorial of", num, "is:", result)
```

```python
#Write a Python function to check whether a number falls within a given range.

def check_range(start, end, num):

    if num >= start and num <= end:
        print(f"{num} falls within the {start}-{end}range.")

    else:
        print(f"{num} does not falls within the {start}-{end} range.")

start_range = int(input("Enter starting range: "))
end_range = int(input("Enter ending range: "))

number = int(input("Enter a number: "))

check_range(start_range, end_range, number)
```

```python
#Write a Python function to Print Even Numbers from a Given List

def even_numbers(numbers):

    print("Even numbers in the list are:")

    for num in numbers:

        if num % 2 == 0:
            print(num)

nums = input("Enter numbers: ")

temp_list = nums.split()

numbers_list = []

for i in temp_list:
    numbers_list.append(int(i))

even_numbers(numbers_list)
```

```python
#Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

def check_prime(num):

    if num <= 1:
        print(num, "is not a prime number.")

    else:

        for i in range(2, num):

            if num % i == 0:
                print(num, "is not a prime number.")
                break

        else:
            print(num, "is a prime number.")

number = int(input("Enter a number: "))

check_prime(number)
```

```python
# Write a Python function that accepts a string and counts the number of upper and lower case letters. 

def count_case(text):

    upper_count = 0
    lower_count = 0

    for ch in text:

        if ch.isupper():
            upper_count += 1

        elif ch.islower():
            lower_count += 1

    print("Number of uppercase letters:", upper_count)
    print("Number of lowercase letters:", lower_count)

string = input("Enter a string: ")
print("input string : ",string)

count_case(string)
```
