```python
import pandas as pd
```

## 1) Practise Pandas Series

#### (i) Create a Pandas Series from Dictionary and Access the elements of a Series in Pandas  


```python
data = {
    "day1": 420,
    "day2": 380,
    "day3": 390
}

ps = pd.Series(data)

print(ps,type(ps))

print(ps["day1"])
print(ps["day3"])
```

#### (ii) Create a Pandas Series from Lists and Access the elements of a Series in Pandas

```python
marks = [85, 90, 78, 88]

ps = pd.Series(marks)

print(ps)

print(ps[0])
print(ps[2])
```

## 2) DataFrames

#### (i) Make a Pandas DataFrame with a two-dimensional Python list 

```python

data = [
    [101, "Mohit", 85],
    [102, "Rahul", 90],
    [103, "Aman", 78]
]

df = pd.DataFrame(data)

print(df)
```

#### (ii) Create DataFrame from Python dict

## 3)Data iteration

#### (i)	Different ways to iterate over rows in Pandas Dataframe 


```python
data = {
    "Name": ["Mohit", "Rahul", "Aman"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

for index, row in df.iterrows():
    print(index, row["Name"], row["Marks"])
```

```python
data = {
    "Name": ["Mohit", "Rahul", "Aman"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)
for row in df.itertuples():
    print(row.Name, row.Marks)
```

#### (ii) Selecting rows in pandas DataFrame based on conditions 

```python
data = {
    "Name": ["Mohit", "Rahul", "Aman"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

res = df[df["Marks"] > 80]

print(res)
```

#### (iii) Select any row from a Dataframe using iloc[]

```python
data = {
    "Name": ["Mohit", "Rahul", "Aman"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print(df.iloc[1])
```

#### (iv) Limited rows selection with given column

```python
data = {
    "Name": ["Mohit", "Rahul", "Aman"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print(df.loc[0:1, ["Name"]])
```

#### (v) Drop rows from the dataframe based on certain condition applied on a column 

```python
data = {
    "Name": ["Mohit", "Rahul", "Aman"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

df = df[df["Marks"] >= 80]

print(df)
```

#### (vi) Insert row at given position in Pandas Dataframe 

```python
data = {
    "Name": ["Mohit", "Rahul"],
    "Marks": [85, 90]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    "Name": ["Aman"],
    "Marks": [78]
})

df = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print(df)
```

#### (vii) Create a list from rows in Pandas dataframe 

```python

data = {
    "Name": ["Mohit", "Rahul", "Aman"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

res = df.values.tolist()

print(res)
print(type(res))
```
