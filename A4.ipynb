```python
# 1) Create a CSV file for address book, CSV file should have column for name, address, mobile, email. Insert 2-3 dummy data entered by user.

import csv
d=[
    ["name", "Address", "mobile", "email"]
    ]


TotalEntries=int(input("Enter total no. of entries you want to insert : "))
for i in range(TotalEntries):
   
   l=[]

   l.append(input("Enter name:"))
   l.append(input("Enter Address:"))
   l.append(int(input("Enter Mobile:")))
   l.append(input("Enter email:"))
   
   d.append(l)
    
with open("Address_book.csv","w",newline='') as file:
    writer = csv.writer(file)

    for row in d:
        writer.writerow(row)
```
