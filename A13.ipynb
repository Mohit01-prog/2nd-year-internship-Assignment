#### 1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]] 

```python
arr = np.array([[6, -8, 73, -110],
               [np.nan, -8, 0, 94]])

print("Original Array:")
print(arr)

arr = np.nan_to_num(arr, nan=0)

print("\nAfter Replacing NaN:")
print(arr)

transposed = arr.T

print("\nAfter Interchanging Rows and Columns:")
print(transposed)
```


#### 2) Move axes of 3D array to new positions

#### 3) Replace NaN values with average of columns 

```python
arr = np.array([[10, np.nan, 30],
                [40, 50, np.nan],
                [70, 80, 90]])

print("Original Array:")
print(arr)

col_mean = np.nanmean(arr, axis=0)

arr = np.nan_to_num(arr, nan=col_mean)

# ======= OR ===============

# inds = np.where(np.isnan(arr))
# arr[inds] = np.take(col_mean, inds[1])

print("\nAfter Replacing NaN with Column Mean:")
print(arr)
```

#### 4) Replace negative value with zero in numpy array using replace 

```python
arr = np.array([5, -2, 8, -7, 10])
print("Original Array:\n",arr)
arr[arr<0] = 0
print("Array after replacing negative values with zeros:\n",arr)
```

#### 5) Study the following program import numpy as np 

#### -> create a numpy 1d-arrays arr1 = np.array([3, 4]) arr2 = np.array([1, 0]) 

#### -> find average of NumPy arrays avg = (arr1 + arr2) / 2 print("Average of NumPy arrays:\n", avg) 

#### -> Calculate average mean median mode values of two NumPy 2d-arrays 

```python
arr1 = np.array([[3,4],
                 [5,7]])

arr2 = np.array([[1,0],
                 [7,8]])
print("Array 1:\n",arr1)
print("\nArray 2:\n",arr2)

avg = (arr1 + arr2)/2

print("Average of NumPy arrays:\n", avg) 

# Combine arrays
combined = np.concatenate((arr1, arr2))

print("\nCombined Array:")
print(combined)

# Mean
mean = np.mean(combined)

# Median
median = np.median(combined)

# Mode
values, counts = np.unique(combined, return_counts=True)
mode = values[np.argmax(counts)]

print("\nMean =", mean)
print("Median =", median)
print("Mode =", mode)
```

#### 6) Solve the following equation using linalg() and inverse Matrix method x - 2y + 3z = 9 -x + 3y - z = -6 2x - 5y + 5z = 17 

```python
A = np.array([[1,-2,3],
              [-1,3,-1],
              [2,-5,5]])

B = np.array([9,-6,17])

print("using linalg():\n")

solution = np.linalg.solve(A,B)
print("solutions:\n",solution,"\n")

print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])

print("\nusing inv():\n")

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print(X,"\n")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])
```
